Chefen vil gerne have en graf der viser salget og specielt profit over tid. Det skal være et linjeplot - med månederne på x-aksen og salget/profit på y-aksen.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Indlæs data fra CSV-filen
data = pd.read_csv('./Data/MatPlotLib-Data/sales.csv')
print(data.head())
````

import pandas as pd
import matplotlib.pyplot as plt

plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.grid(True)

plt.show()

### Fordeling af salg af hvert produkt i en piechart

Chefen vil gerne have en graf der viser salget af hvert produkt i en piechart. Det skal være en piechart - med produkterne på x-aksen og salget af hvert produkt på y-aksen.


````python
if data is not None:
    print("Data er loaded korrekt og er klar til at lave et cirkeldiagram")
else:
    print("Data er ikke loaded korrekt")
````

Følgende er en vejledning til at løse opgaven.

<details>
<summary>Klik her for løsning</summary>

```python
# Beregn det totale salg for hvert produkt
total_sales = {
    'Face Cream': data['facecream'].sum(),
    'Face Wash': data['facewash'].sum(),
    'Toothpaste': data['toothpaste'].sum(),
    'Bathing Soap': data['bathingsoap'].sum(),
    'Shampoo': data['shampoo'].sum(),
    'Moisturizer': data['moisturizer'].sum()
}

# Opret et pie chart for fordelingen af salg af hvert produkt
plt.figure(figsize=(8, 8))
plt.pie(total_sales.values(), labels=total_sales.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution of Different Products')
plt.axis('equal')  # Sikrer, at pie chartet er cirkulært
plt.show()
```



### Lav en oversigt med et "stack plot"

Chefen synes stadig det er en smule uoverskueligt og vil gerne have du lavet et "stack plot" med de forskellige produkt kategorier. 

````python
if data is not None:
    print("Data er loaded korrekt og er klar til at lave et stack plot")
else:
    print("Data er ikke loaded korrekt")
````

Følgende er en vejledning til at løse opgaven.

<details>
<summary>Klik her for løsning</summary>

```python
# Opret et stack plot for salget af forskellige produkter over måneder
plt.figure(figsize=(12, 8))

# Definer data til stack plot
months = data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']

# Opret stack plot
plt.stackplot(months, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer, 
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.title('Stack Plot of Product Sales Over Time')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(months)
plt.legend(loc='upper left')
plt.show()
```


## 3. Vigtige plots inden for Machine Learning! - ML resultater

````python
import pandas as pd
import matplotlib.pyplot as plt

# Indlæs data fra CSV-filen
data = pd.read_csv('./Data/MatPlotLib-Data/DKHousingPricesSample100k.csv')

# Fjern rækker med manglende værdier i 'sqm'
data = data[data['sqm'].notna()]

# Fjern 'house_id' kolonnen
data = data.drop("house_id", axis=1)

# Beregn gennemsnitlig sqm_price for hver region
average_sqm_price_by_region = data.groupby('region')['sqm_price'].mean()

# Visualiser det gennemsnitlige sqm_price for hver region
plt.figure(figsize=(10, 6))
average_sqm_price_by_region.plot(kind='bar', color='skyblue')
plt.title('Average Square Meter Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Square Meter Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
````
